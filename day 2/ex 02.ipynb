```python
import pandas as pd
```

```python
dataset = pd.read_csv("dataset.csv")
```

```python
reg_name = dataset["reg_name"]
amount = dataset["amount"]
year = dataset["year"]
```

```python
total_year = list()

for i in year:
    if i not in total_year:
        total_year.append(i)
total_year = sorted(total_year)
```

```python
total = dict()

for i in reg_name:
    if i not in total.keys():
        total[i] = []
        for n in total_year:
            if n not in total[i]:
                total[i].append([n, 0])
```

```python
total_amount = list()

for i in total.keys():
    k = 0
    for n in reg_name:
        if n == i:
            t = 0
            for h in total_year:
                if h == year[k]:
                    total[i][t][1] += amount[k]
                t += 1
        k += 1
```

```python
for i in total.keys():
    nb = 0
    print(i)
    for n in total[i]:
        print(total[i][nb][0], "->", total[i][nb][1], "euros")
        nb += 1
```
